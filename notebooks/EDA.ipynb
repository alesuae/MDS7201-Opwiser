```python
import pandas as pd
from ydata_profiling import ProfileReport
import matplotlib.pyplot as plt
import os
```

```python
os.chdir('../')
os.getcwd()
```

```python
df = pd.read_csv('processed/dataset.csv')
df
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))
data_combined = df.copy()
# Gráfico de barras para las ventas
ax1.bar(data_combined['fecha'], data_combined['venta_total_neto'], color='skyblue', label='Ventas', alpha=0.7)
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Ventas', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Crear un segundo eje y para el stock
ax2 = ax1.twinx()
ax2.plot(data_combined['fecha'], data_combined['stock_disponible_total'], color='orange', label='Stock', linewidth=2)
ax2.set_ylabel('Stock', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Agregar leyenda
fig.legend(loc='upper right', bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)

# Título y estilo del gráfico
plt.title('Ventas y Stock a lo Largo del Tiempo')
plt.tight_layout()
plt.show()
```
