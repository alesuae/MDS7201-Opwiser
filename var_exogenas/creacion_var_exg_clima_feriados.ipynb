# Fechas Festivas

```python
import pandas as pd
import holidays
from holidays import country_holidays
from datetime import datetime

# Definir los años que te interesan
years = range(2018, 2025)  # Del 2018 al 2024 inclusive

# Obtener los feriados para Chile en esos años
chile_holidays = country_holidays('CL', years=years)

# Extraer solo las fechas (las claves del diccionario)
fechas_feriados = list(chile_holidays.keys())

# Crear un DataFrame con las fechas
feriados_cl = pd.DataFrame(fechas_feriados, columns=['fecha'])
feriados_cl['fecha'] = pd.to_datetime(feriados_cl['fecha'], format='%Y-%m-%d')

# Mostrar el DataFrame resultante
feriados_cl.head()
```

```python
# Para agregarlo al df final del modelo

# Crear una nueva columna 'es_festivo' en df_stock, toma valor 1 si es feriado y 0 si no
df_2.loc[:, 'es_festivo'] = df_2['fecha'].isin(feriados_cl['fecha']).astype(int)

df_2
```

```python
from meteostat import Point, Daily

# Por simplicidad utilizaremos sólo datos de Santiago
santiago = Point(-33.45694, -70.64827) # https://www.geodatos.net/coordenadas/chile/santiago

datos_clima = Daily(santiago, datetime(2018, 3, 1), datetime(2024, 6, 30))
datos_clima = datos_clima.fetch()

# Arreglar 'índice' fecha para que sea columna
datos_clima.reset_index(inplace=True)
datos_clima.rename(columns={'index': 'fecha'}, inplace=True)

datos_clima.rename(columns={'time': 'fecha'}, inplace=True)
datos_clima['fecha'] = pd.to_datetime(datos_clima['fecha'])

datos_clima.head()
# tavg: Temperatura promedio (°C)
# tmin: Temperatura mínima (°C)
# tmax: Temperatura máxima (°C)
# prcp: Precipitación (ml)
# snow: Nieve (ml)
# wdir: Dirección del viento (°)
# wspd: Velocidad del viento (m/s)
# wpgt: Racha máxima de viento (m/s)
# pres: Presión atmosférica (hPa)
# tsun: Horas de sol
```

```python
datos_clima.dtypes
```

```python
datos_clima.isnull().sum()
# Eliminamos snow, wdir, wpgt y tsun
```

```python
datos_clima.drop(columns=['snow', 'wdir', 'wpgt', 'tsun'], inplace=True)
datos_clima.head()
```
